### read in and clean target table

```python
# define the file paths
file_path_male = '/Users/roger.qiu/Documents/other/code/USD/12. Capstone Project/ADS-599-Capstone/module_1/option2/IHME_USA_STOMACH_CANCER_COUNTY_RACE_ETHN_2000_2019_MX_MALE/IHME_USA_STOMACH_CANCER_COUNTY_RACE_ETHN_2000_2019_MX_2019_MALE_Y2023M06D28.CSV'

file_path_female = '/Users/roger.qiu/Documents/other/code/USD/12. Capstone Project/ADS-599-Capstone/module_1/option2/IHME_USA_STOMACH_CANCER_COUNTY_RACE_ETHN_2000_2019_MX_FEMALE/IHME_USA_STOMACH_CANCER_COUNTY_RACE_ETHN_2000_2019_MX_2019_FEMALE_Y2023M06D28.CSV'

# read in csv file
df_male = pd.read_csv(file_path_male)

df_male.head()
```

```python
# read in csv file
df_female = pd.read_csv(file_path_female)

df_female.head()
```

```python
# union male and female datasets together
df_combined = pd.concat([df_female, df_male], ignore_index=True)
df_combined.head()
```

```python
df_combined.shape
```

```python
# check distinct values, if only one, can drop
df_combined['cause_name'].unique()
```

```python
# select important fields
df_select = df_combined[['location_id', 'fips', 'race_name', 'sex_name', 'age_name', 'val']]
df_select.head()
```

```python
df_select.shape
```

```python
# filter out rows where 'fips' is null, we'll need this field
df_filtered = df_select.dropna(subset=['fips'])

# also filter out rows where 'race_name' is 'Total'
df_filtered = df_filtered[df_filtered['race_name'] != 'Total']

df_filtered.head()
```
